# Regresión Logística

Este Jupyter Notebook implementa un modelo de Regresión Logística,
utilizado en minería de datos para problemas de clasificación supervisada.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# Cargar el conjunto de datos
X, y = load_iris(return_X_y=True)

# Crear y entrenar el modelo
modelo = LogisticRegression(max_iter=200)
modelo.fit(X, y)

# Realizar predicciones
predicciones = modelo.predict(X)

predicciones[:4]
```

### Explicación del Notebook

La regresión logística es un algoritmo de clasificación supervisada que
calcula la probabilidad de que una observación pertenezca a una clase
determinada. En este notebook se entrena el modelo y se realizan predicciones
para comprobar su funcionamiento.
