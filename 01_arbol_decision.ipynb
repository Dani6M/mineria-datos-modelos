# Árbol de Decisión

Este Jupyter Notebook implementa un modelo de Árbol de Decisión,
un algoritmo de clasificación supervisada utilizado en minería de datos.


from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

modelo = DecisionTreeClassifier()
modelo.fit(X, y)

predicciones = modelo.predict(X)
predicciones[:5]


### Explicación

Este ejemplo muestra cómo entrenar un Árbol de Decisión
y realizar predicciones utilizando Python y scikit-learn.
