# ARIMA – Series de Tiempo

Este Jupyter Notebook implementa el modelo ARIMA,
utilizado para el análisis y pronóstico de series de tiempo.


```python
from statsmodels.tsa.arima.model import ARIMA

# Serie de tiempo de ejemplo
data = [10, 12, 14, 16, 18, 20,22,24,26]

# Crear y entrenar el modelo ARIMA
modelo = ARIMA(data, order=(1, 1, 1))
modelo_entrenado = modelo.fit()

# Realizar pronóstico
pronostico = modelo_entrenado.forecast(steps=2)
pronostico
```

### Explicación del Notebook

El modelo ARIMA es un algoritmo supervisado utilizado
para analizar series de tiempo. Considera la dependencia
entre observaciones pasadas para realizar pronósticos
de valores futuros.
