# Regresión Lineal

Este notebook implementa un modelo de Regresión Lineal,
utilizado para predecir valores numéricos continuos.


```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Datos de ejemplo
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Realizar una predicción
prediccion = modelo.predict([[5]])
prediccion
```

### Explicación del Notebook

La regresión lineal establece una relación entre variables
independientes y una variable dependiente, permitiendo
realizar predicciones numéricas.
